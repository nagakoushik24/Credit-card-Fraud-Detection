## Train Test and Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

## Decision Tress

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
dtree.fit(X_train,y_train)
```

```python
dtree.fit(X_train,y_train)
```

## Prediction and Evaluation

```python
predictions = dtree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

# Assuming 'dtree' is your Decision Tree model and 'X' is the feature matrix
plt.figure(figsize=(20, 25))
tree.plot_tree(
    dtree,
    feature_names=X.columns.tolist(),  # Convert Index to list
    class_names=['Class-1', 'Class-0'],  # Class labels
    rounded=True,  # Rounded node edges
    filled=True,  # Adds color according to class
    proportion=True  # Show proportions at nodes
)
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Assuming 'predictions' are the model predictions on 'y_test'
print("Classification Report:")
print(classification_report(y_test, predictions))

# Calculate and print accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Assuming 'y_test' are the true labels and 'predictions' are the model's predictions
conf_matrix = confusion_matrix(y_test, predictions)

# Print the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)

# Plot the confusion matrix using Seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False, 
            xticklabels=['Class-0', 'Class-1'], yticklabels=['Class-0', 'Class-1'])

# Add labels and title
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
